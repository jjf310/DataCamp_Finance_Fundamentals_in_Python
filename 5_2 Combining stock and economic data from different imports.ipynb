```python
#Combining stock and economic data
import pandas as pd
from datetime import date
from pandas_datareader.data import DataReader
from pandas.util.testing import assert_frame_equal
import matplotlib.pyplot as plt
```

```python
#start date
start = date(2001,1,1 )
#series WTI
series = 'DCOILWTICO'
#Datareader
oil= DataReader(series,'fred', start)
#stockticker Chevron
ticker='CVX'
#Datareader
stock= DataReader(ticker,'yahoo', start)
#Concat horizontally using axis=1, pandas will aline the data on the index so the dates match
data = pd.concat([stock[['Close']], oil], axis=1)
```

```python
data.info()
```

```python
#rename the two columns
data.columns = ['Chevron', 'Oil Price']
data.plot()
plt.show()
```
